```python
import datetime as dt
from scipy.interpolate import make_interp_spline
from scipy.interpolate import interp1d
import numpy as np
from datetime import time
from random import random





def pulse_attack(df,dd,mm,yy,attack_point_time,pulse_parameter):
    row_num = []
    new_date=dt.date(yy,mm,dd)
    random_parameter=0
    for i in df.index:
        if(df['Date'][i]==new_date ):
            if(df['Hour'][i]==attack_point_time):
                df['SystemLoad'][i]=df['SystemLoad'][i]*(1+pulse_parameter)
                details=[]
                details.append(df['Date'][i])
                details.append(df['Hour'][i])
                row_num.append(details)                

    print("Random Attack succesfully done on following data points : ")
    print(row_num)
```

```python
from datetime import timedelta, datetime
from random import random
import random

starting_date = datetime.strptime("2004-1-1", "%Y-%m-%d")

ending_date = datetime.strptime("2008-12-31", "%Y-%m-%d")

current_date=starting_date

while(current_date<=ending_date):
    
    dd=current_date.day
    mm=current_date.month
    yy=current_date.year
    
    num1=random.randint(1,24)

    
    print("Loop")
    print(df)
    print(dd)
    print(mm)
    print(yy)
    print(num1)

    pulse_attack(df,dd,mm,yy,num1,0.5)
    current_date=current_date+timedelta(1)



print("Done Attacking")
```
